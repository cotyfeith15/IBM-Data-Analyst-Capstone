### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
import pandas as pd
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip()

# Group by Age and calculate the median ConvertedCompYearly
median_compensation_by_age = df.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(median_compensation_by_age['Age'], median_compensation_by_age['ConvertedCompYearly'], marker='o')

# Adding title and labels
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)  # Rotate age group labels for better readability

plt.grid()  # Add grid for easier reading
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Define a mapping for the age buckets
age_mapping = {
    'Under 18 years old': 0,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    'Over 64 years old': 70
}

# Map the age buckets to their corresponding numeric values
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Filter the DataFrame for relevant age buckets (25-34 and 35-44)
filtered_df = df[df['Age'].isin(['25-34 years old', '35-44 years old'])]

# Group by Age and calculate the median ConvertedCompYearly
median_compensation_filtered = filtered_df.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(median_compensation_filtered['Age'], median_compensation_filtered['ConvertedCompYearly'], marker='o')

# Adding title and labels
plt.title('Median Yearly Compensation for Age Groups 25-34 and 35-44')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)  # Rotate age group labels for better readability

plt.grid()  # Add grid for easier reading
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
import pandas as pd
import matplotlib.pyplot as plt

experience_column = 'WorkExp' 

# Group by experience level and calculate the average job satisfaction
job_satisfaction_by_experience = df.groupby(experience_column)['JobSatPoints_6'].mean().reset_index()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(job_satisfaction_by_experience[experience_column], job_satisfaction_by_experience['JobSatPoints_6'], marker='o')

# Adding title and labels
plt.title('Job Satisfaction by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Job Satisfaction (JobSatPoints_6)')
plt.xticks(rotation=45)  # Rotate experience level labels for better readability

plt.grid()  # Add grid for easier reading
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
