### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Function to split and count languages
def count_languages(column):
    languages = []
    for row in df[column].dropna():
        languages.extend([lang.strip() for lang in row.split(";")])
    return Counter(languages)

# Count frequencies
worked_with_counts = count_languages("LanguageHaveWorkedWith")
want_to_work_with_counts = count_languages("LanguageWantToWorkWith")

# Create a DataFrame for visualization
lang_df = pd.DataFrame([
    {"Language": lang, "Category": "Worked With", "Count": worked_with_counts[lang]}
    for lang in worked_with_counts
] + [
    {"Language": lang, "Category": "Want to Work With", "Count": want_to_work_with_counts[lang]}
    for lang in want_to_work_with_counts
])

# Optional: Filter for the top N languages for clarity (e.g., top 10)
top_langs = lang_df.groupby("Language")["Count"].sum().nlargest(10).index
lang_df = lang_df[lang_df["Language"].isin(top_langs)]

# Create the grouped bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Language", y="Count", hue="Category", data=lang_df, palette=["blue", "orange"])
plt.xticks(rotation=45)
plt.title("Comparison of Programming Languages Used vs. Desired")
plt.ylabel("Count")
plt.xlabel("Programming Language")
plt.legend(title="Category")
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
import pandas as pd

# Create a DataFrame that counts RemoteWork responses by Region
remote_work_counts = df.groupby(['Country', 'RemoteWork']).size().unstack(fill_value=0)

# Display the counts (optional)
print(remote_work_counts)

import matplotlib.pyplot as plt
import seaborn as sns

# Create the grouped bar chart
remote_work_counts.plot(kind='bar', figsize=(200, 90), color=['skyblue', 'salmon'])
plt.title("Distribution of Remote Work by Country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Remote Work Status")
plt.show()

# Create the heatmap
plt.figure(figsize=(200,90))
sns.heatmap(remote_work_counts, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title("Heatmap of Remote Work Distribution by Country")
plt.xlabel("Remote Work Status")
plt.ylabel("Country")
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
import pandas as pd

# Sample DataFrame
# df = pd.read_csv("your_dataset.csv")  # Load your DataFrame

# Step 1: Ensure JobSat is numeric
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')

# Step 2: Convert YearsCodePro to numeric
def convert_years_code(value):
    # Check if value is a string, otherwise return the float value
    if isinstance(value, str):
        if "Less than 1" in value:
            return 0
        elif "More than" in value:
            return float(value.split()[2])  # Extract the number
        else:
            return float(value)  # If it's already a numeric string
    else:
        return value  # Return the value as is if it's already a float or NaN

df['YearsCodePro'] = df['YearsCodePro'].apply(convert_years_code)

# Step 3: Handle NaN values in JobSat
df['JobSat'] = df['JobSat'].fillna(df['JobSat'].mean())

# Optional: Drop NaN rows
df_cleaned = df.dropna(subset=['JobSat', 'YearsCodePro'])

# Step 4: Calculate correlation
pearson_corr = df_cleaned[['JobSat', 'YearsCodePro']].corr(method='pearson')
print("Pearson Correlation:\n", pearson_corr)

spearman_corr = df_cleaned[['JobSat', 'YearsCodePro']].corr(method='spearman')
print("Spearman Correlation:\n", spearman_corr)
```
